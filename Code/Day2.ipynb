Data preprocessing (Importing useful libraries, Normalizing data, Splitting into train and test)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/home/aditya/Documents/100DayofMLcode/Datasets/Day2.csv')
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler() 
df = scaler.fit_transform(df)
```

```python
X = df[:,:1]
Y = df[:,1]
```

```python
print(X.shape)
print(Y.shape)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split( X , Y , test_size = 0.2, random_state = 0)
```

Importing Linear Regression model from Sklearn

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor = regressor.fit(X_train, Y_train)
```

```python
Y_pred = regressor.predict(X_test)
X_pred = regressor.predict(X_train)
```

Plotting the Train and Test Result

```python
plt.scatter(X_train , Y_train, color = 'red')
plt.plot(X_train , X_pred, color ='blue')
plt.show()
```

```python
plt.scatter(X_test , Y_test, color = 'red')
plt.plot(X_test , Y_pred, color ='blue')
plt.show()
```
